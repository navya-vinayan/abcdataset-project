```python
team=df["Team"].value_counts()
percentage=(team/len(df))*100
result=pd.DataFrame({"Employees":team,"Percentage":percentage})
print(result)
```

```python
team.plot(kind="bar",figsize=(10,5))
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.title("Distribution of employees across each team",color="r")
plt.show()
```

Observation
The New Orleans Pelicans have the highest number of employees, followed by the Memphis Grizzlies. The Orlando Magic and Minnesota Timberwolves have the fewest employees. Overall, the number of employees across teams is fairly similar, with only minor differences between them.



Observation: Employees are divided into 5 position.SG has Highest number and C has the lowest.


```python
bins = [18,21,24,27,30,33,36,39,42,45]
df["Age Group"]=pd.cut(df['Age'],bins=bins)
age_group=df["Age Group"].value_counts().sort_index()
print(age_group)
print("Predominant Age Group:",age_group.idxmax())


```

```python
age_group.plot(kind="bar")
plt.title("Distribution of employees across each age group",color="pink")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.show()
```

The majority of employees are between 24 and 27 years old. As age increases, the number of employees gradually declines. There are no employees in the 42–45 age group.


```python
team_salary=df.groupby("Team")["Salary"].sum()
print(team_salary)
print("Highest Salary Team:",team_salary.idxmax())
```

```python
position_salary=df.groupby("Position")["Salary"].sum()
print(position_salary)
print("Highest Salary Position:",position_salary.idxmax())
```

```python
team_salary.plot(kind="bar", figsize=(12,5))
plt.title("Salary Expenditure by Team")
plt.xlabel("Team")
plt.ylabel("Total Salary")
plt.show()
```



Observations: Cleveland Cavaliers have the highest salary.Philadelphia 76ers have the lowest salary expenditure


```python
position_salary.plot(kind="bar")
plt.title("Salary Expenditure by Position")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

Observation: C have the highest total salary expenditure. SG have the lowest salary expenditure.

```python
corr=df["Age"].corr(df["Salary"])
print("Correlation:",corr)
```

```python
plt.scatter(df["Age"],df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```

Observation: There is a slight positive correlation between age and salary. Salaries differ considerably among employees within the same age group, and a few employees receive exceptionally high salaries.
